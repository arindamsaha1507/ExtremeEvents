```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as clrs
import os
from tqdm import tqdm
```

```python
directory = '../Cluster/Data/Raw/Data/'
plot_data = []

for file in tqdm(os.listdir(directory)):

    full_file_name = directory + file
    data = np.loadtxt(full_file_name)
    assert len(np.unique(data[:,1])) == 1
    y_point = data[0,1]

    for x_point in np.unique(data[:,0]):
        point_color = max(data[data[:,0]==x_point,4])
        
        plot_data.append(np.array([x_point, y_point, point_color]))

plot_data = np.array(plot_data)
```

```python
plt.figure(figsize=(16,16))
plt.scatter(plot_data[:,0], plot_data[:,1], c=plot_data[:,2], s=16, marker='.')
plt.xlim((0,4000))
plt.ylim((0,0.2))
plt.colorbar()
plt.show()
```

```python
directory = '../Cluster/Data/Raw/Data/'
plot_data = []

for file in tqdm(os.listdir(directory)):

    full_file_name = directory + file
    data = np.loadtxt(full_file_name)
    assert len(np.unique(data[:,1])) == 1
    y_point = data[0,1]

    for x_point in np.unique(data[:,0]):
        
        sliced_xp = data[data[:,0]==x_point]
        
        point_color1 = max(sliced_xp[sliced_xp[:,2]==0,4])
        point_color2 = max(sliced_xp[sliced_xp[:,2]==1,4])
                
        plot_data.append(np.array([x_point, y_point, point_color1, point_color2]))

plot_data = np.array(plot_data)
```

```python
fig,ax = plt.subplots(1,2,sharey=True, figsize=(16,9))

ax[0].scatter(plot_data[:,0], plot_data[:,1], c=plot_data[:,2], s=16, marker='.')
ax[1].scatter(plot_data[:,0], plot_data[:,1], c=plot_data[:,3], s=16, marker='.')

ax[0].set_xlim((100,4000))
ax[1].set_xlim((100,4000))
ax[0].set_ylim((0,0.2))

plt.tight_layout()
plt.show()
```

```python
directory = '../Cluster/Data/Raw/Data/'
plot_data = []

for file in tqdm(os.listdir(directory)):

    full_file_name = directory + file
    data = np.loadtxt(full_file_name)
    assert len(np.unique(data[:,1])) == 1
    y_point = data[0,1]

    for x_point in np.unique(data[:,0]):
        
        sliced_xp = data[data[:,0]==x_point]
        
        sliced_xp1 = sliced_xp[sliced_xp[:,2]==0]
        sliced_xp2 = sliced_xp[sliced_xp[:,2]==1]
        
        point_color1 = len(sliced_xp1[sliced_xp1[:,4]>1])
        point_color2 = len(sliced_xp2[sliced_xp2[:,4]>1])
                
        plot_data.append(np.array([x_point, y_point, point_color1, point_color2]))

plot_data = np.array(plot_data)
```
