```python
import numpy as np
import matplotlib.pyplot as plt
import os
from tqdm import tqdm
```

```python
fname = '../Cluster/Data/Raw/bif_data_00001_00100.dat'
data = np.loadtxt(fname)
```

```python
assert len(np.unique(data[:,1])) == 1
y_point = data[0,1]
print(y_point)
```

```python
directory = '../Cluster/Data/Raw/'
plot_data = []

for file in tqdm(os.listdir(directory)):

    full_file_name = directory + file
    data = np.loadtxt(full_file_name)
    assert len(np.unique(data[:,1])) == 1
    y_point = data[0,1]

    for x_point in np.unique(data[:,0]):
        point_color = max(data[data[:,0]==x_point,4])
        
        plot_data.append(np.array([x_point, y_point, point_color]))

plot_data = np.array(plot_data)
```

```python
len(os.listdir(directory))
```

```python
np.array(plot_data)
```

```python
plt.figure(figsize=(16,16))
plt.scatter(plot_data[:,0], plot_data[:,1], c=plot_data[:,2], s=64, marker='s')
plt.xlim((0,4000))
plt.ylim((0,0.2))
plt.show()
```
